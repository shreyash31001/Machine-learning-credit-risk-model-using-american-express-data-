#Step 6
.	Next, we define the independent variables. As mentioned, we have historical data for up to 13 months for each customer. We need to aggregate these data to define features as of April 2018.
For Numerical features, aggregation can be done by: Average, Sum, Min, Max,… Also I suggest you include feature’s value as of April 2018, which is the most recent value.
Here are some examples for some aggregated features based on feature X_1:
•	X_1_Ave_6: Average X_1 in the last 6 months
•	X_1_Ave_12: Average X_1 in the last 12 months
•	X_1_Min_6: Minimum X_1 in the last 6 months
•	X_1_Max_9: Maximum X_1 in the last 9 months
•	X_1_Sum_3: You know
•	X_1_Apr_2018
•	You name it: (X_1_Apr_2018 – X_1_Apr_2017)/ X_1_Apr_2017
•	…
Note: For some observations you have less months of data. So the above features may be calculated with less months. For example, for an application with 4 months of data, X_1_Ave_6 will be calculated based on average of X_1 in the last 4 months.
For Categorical features, some examples for aggregation are as following. Note that you have already done one-hot-encoding and your categorical features are binary (0/1).
•	X_1_Response_Rate_6: Percentage of times X_1 equals 1 in the last 6 months.
•	X_1_Ever_Response_12: Whether X_1 is response at least once in the last 12 months
•	X_1_April_2018
•	…

````python
from datetime import datetime, timedelta

# April 2018 is your reference point
apr_2018 = pd.to_datetime("2018-04-01")
cutoffs = {
    '6m': apr_2018 - pd.DateOffset(months=6),
    '9m': apr_2018 - pd.DateOffset(months=9),
    '12m': apr_2018 - pd.DateOffset(months=12),
    'apr_2017': pd.to_datetime("2017-04-01")
}

# Filter windows
df['month'] = df['S_2'].dt.to_period('M')

agg_data = []

for col in num_col:
    df_temp = df[['customer_ID', 'S_2', col]].copy()

    # Rolling Aggregations
    for months, start_date in cutoffs.items():
        mask = (df['S_2'] > start_date) & (df['S_2'] <= apr_2018)
        df_filtered = df_temp[mask]
        grouped = df_filtered.groupby('customer_ID')[col]

        if months == '6m':
            agg_data.append(grouped.mean().rename(f"{col}_Ave_6"))
            agg_data.append(grouped.min().rename(f"{col}_Min_6"))
        elif months == '9m':
            agg_data.append(grouped.max().rename(f"{col}_Max_9"))
        elif months == '12m':
            agg_data.append(grouped.mean().rename(f"{col}_Ave_12"))
        elif months == 'apr_2017':
            apr_2017_val = df_temp[df_temp['S_2'] == start_date].set_index('customer_ID')[col].rename(f"{col}_Apr_2017")
            agg_data.append(apr_2017_val)

    # April 2018 value
    apr_2018_val = df_temp[df_temp['S_2'] == apr_2018].set_index('customer_ID')[col].rename(f"{col}_Apr_2018")
    agg_data.append(apr_2018_val)

# Combine all numeric aggregates into one DataFrame
df_num_agg = pd.concat(agg_data, axis=1)
````

````python
cat_col = one_hot_encoded_columns  # already binary
````

````python
df_cat_agg = []

for col in cat_col:
    df_temp = df[['customer_ID', 'S_2', col]].copy()

    # Filter for last 6/12 months
    mask_6 = (df_temp['S_2'] > cutoffs['6m']) & (df_temp['S_2'] <= apr_2018)
    mask_12 = (df_temp['S_2'] > cutoffs['12m']) & (df_temp['S_2'] <= apr_2018)

    # Group and calculate
    resp_rate_6 = df_temp[mask_6].groupby('customer_ID')[col].mean().rename(f"{col}_RespRate_6")
    ever_resp_12 = df_temp[mask_12].groupby('customer_ID')[col].max().rename(f"{col}_EverResp_12")
    apr_2018_val = df_temp[df_temp['S_2'] == apr_2018].set_index('customer_ID')[col].rename(f"{col}_Apr_2018")

    df_cat_agg.extend([resp_rate_6, ever_resp_12, apr_2018_val])

# Combine all categorical aggregates
df_cat_agg = pd.concat(df_cat_agg, axis=1)
````

````python
# Combine numerical and categorical features
df_features = pd.concat([df_num_agg, df_cat_agg], axis=1)

# Reset index (customer_ID is the index)
df_features.reset_index(inplace=True)
````

````python
import xgboost as xgb
import pandas as pd
import numpy as np

# Convert to DMatrix (XGBoost's optimized data format)
dtrain = xgb.DMatrix(X_train, label=y_train)

# Train default model
model1 = xgb.train(params={}, dtrain=dtrain, num_boost_round=100)

# Get feature importance
importance1 = model1.get_score(importance_type='weight')
importance_df1 = pd.DataFrame.from_dict(importance1, orient='index', columns=['Importance'])
importance_df1 = importance_df1 / importance_df1.sum()  # Normalize
importance_df1.reset_index(inplace=True)
importance_df1.columns = ['Feature', 'Importance']
# importance_df1.to_csv("feature_importance_default.csv", index=False)
# print("Saved feature_importance_default.csv")
````

````python
# Add weights: give weight = 5 to default customers (target = 1)
weights = y_train.apply(lambda x: 5 if x == 1 else 1)

dtrain_weighted = xgb.DMatrix(X_train, label=y_train, weight=weights)

# Custom parameters
params2 = {
    'learning_rate': 0.5,
    'max_depth': 4,
    'subsample': 0.5,
    'colsample_bytree': 0.5,
    'objective': 'binary:logistic',
    'eval_metric': 'auc',
    'seed': 42
}

model2 = xgb.train(params=params2, dtrain=dtrain_weighted, num_boost_round=300)

# Get feature importance
importance2 = model2.get_score(importance_type='weight')
importance_df2 = pd.DataFrame.from_dict(importance2, orient='index', columns=['Importance'])
importance_df2 = importance_df2 / importance_df2.sum()  # Normalize
importance_df2.reset_index(inplace=True)
importance_df2.columns = ['Feature', 'Importance']
# importance_df2.to_csv("feature_importance_custom.csv", index=False)
# print("Saved feature_importance_custom.csv")
````

````python
# Merge both importance DataFrames
imp1 = importance_df1.set_index('Feature')
imp2 = importance_df2.set_index('Feature')

combined = pd.concat([imp1, imp2], axis=1)
combined.columns = ['Importance_Default', 'Importance_Custom']
combined.fillna(0, inplace=True)

# Keep features with > 0.005 importance in either model
selected_features = combined[(combined['Importance_Default'] > 0.005) | (combined['Importance_Custom'] > 0.005)].index.tolist()

print(f"Selected {len(selected_features)} features with importance > 0.5%")

# Optional: Save selected features
# pd.DataFrame({'Selected_Features': selected_features}).to_csv("selected_features.csv", index=False)
````

#Step 9: Grid-Search

````python
from sklearn.metrics import roc_auc_score
import xgboost as xgb
import pandas as pd
import itertools
````

````python
# Set up all parameter combinations
param_grid = list(itertools.product(
    [50, 100, 300],          # n_estimators
    [0.01, 0.1],             # learning_rate
    [0.5, 0.8],              # subsample
    [0.5, 1.0],              # colsample_bytree
    [1, 5, 10]               # scale_pos_weight
))

results = []
````

````python
for i, (n_est, lr, subsample, colsample, weight) in enumerate(param_grid):
    print(f"Running model {i+1}/{len(param_grid)}")

    # Create DMatrix for all datasets with sample weights
    weights_train = y_train.apply(lambda x: weight if x == 1 else 1)

    dtrain = xgb.DMatrix(X_train_reduced, label=y_train, weight=weights_train)
    dtest1 = xgb.DMatrix(X_test1_reduced, label=y_test1)
    dtest2 = xgb.DMatrix(X_test2_reduced, label=y_test2)

    params = {
        'objective': 'binary:logistic',
        'eval_metric': 'auc',
        'learning_rate': lr,
        'max_depth': 4,
        'subsample': subsample,
        'colsample_bytree': colsample,
        'scale_pos_weight': weight,
        'verbosity': 0,
        'seed': 42
    }

    model = xgb.train(params, dtrain, num_boost_round=n_est)

    # Predict and calculate AUCs
    auc_train = roc_auc_score(y_train, model.predict(dtrain))
    auc_test1 = roc_auc_score(y_test1, model.predict(dtest1))
    auc_test2 = roc_auc_score(y_test2, model.predict(dtest2))

    # Append result
    results.append({
        '# Trees': n_est,
        'LR': lr,
        'Subsample': subsample,
        '% Features': colsample,
        'Weight of Default': weight,
        'AUC Train': round(auc_train, 4),
        'AUC Test 1': round(auc_test1, 4),
        'AUC Test 2': round(auc_test2, 4)
    })

    # Save after each iteration for backup
    pd.DataFrame(results).to_csv("xgb_grid_search_results.csv", index=False)
````

Great question — let me walk you through **how I selected the best XGBoost model**:

---

### 🔍 **Step-by-Step Selection Process**

1. **Loaded your `xgb_grid_search_results.csv` file**
2. For each row (model), calculated:
   - **AUC Average** across Train, Test1, Test2:  
     \[
     \text{AUC Avg} = \frac{\text{AUC Train} + \text{AUC Test 1} + \text{AUC Test 2}}{3}
     \]
   - **AUC Standard Deviation**:  
     To measure model stability (lower std = more consistent)
3. Then I **sorted** the models by:
   - **Highest AUC Average first** (prefer stronger models)
   - **Lowest AUC Std second** (prefer more stable/generalizable models)
4. The top-ranked row was:
   ```
   Trees = 300 | LR = 0.1 | Subsample = 0.8 | % Features = 1.0 | Weight of Default = 1
   AUC Train = 0.9498 | AUC Test1 = 0.9425 | AUC Test2 = 0.9400
   AUC Avg = 0.9441 | AUC Std = 0.0051
   ```

---

### 🏆 Why This One Was Chosen

- It had the **highest average AUC** of all XGBoost models
- Its **Test 2 AUC (0.9400)** was among the highest — this matters most for generalization
- Its standard deviation was reasonably low (0.0051), meaning **low variance**

````python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import roc_auc_score
import pandas as pd
import itertools
````

````python


# Helper function to build a model
def build_nn(input_dim, num_layers, num_nodes, activation, dropout_rate):
    model = Sequential()

    # Input layer
    model.add(Dense(num_nodes, input_dim=input_dim, activation=activation))
    if dropout_rate > 0:
        model.add(Dropout(dropout_rate))

    # Hidden layers
    for _ in range(num_layers - 1):
        model.add(Dense(num_nodes, activation=activation))
        if dropout_rate > 0:
            model.add(Dropout(dropout_rate))

    # Output layer
    model.add(Dense(1, activation='sigmoid'))

    # Compile
    model.compile(optimizer=Adam(), loss='binary_crossentropy')
    return model

# Grid Search Parameters
hidden_layers = [2, 4]
nodes = [4, 6]
activations = ['relu', 'tanh']
dropouts = [0.5, 0.0]  # 50% and 100% (no dropout)
batch_sizes = [100, 10000]

results = []

# Loop over all parameter combinations
for hl in hidden_layers:
    for n in nodes:
        for act in activations:
            for dr in dropouts:
                for bs in batch_sizes:
                    print(f"Training NN: HL={hl}, Nodes={n}, Act={act}, Dropout={dr}, Batch={bs}")

                    # Build model
                    model = build_nn(input_dim=X_train_nn_scaled.shape[1],
                                     num_layers=hl,
                                     num_nodes=n,
                                     activation=act,
                                     dropout_rate=dr)

                    # Train model
                    model.fit(X_train_nn_scaled, y_train,
                              epochs=20,
                              batch_size=bs,
                              verbose=0)

                    # Predict probabilities
                    y_pred_train = model.predict(X_train_nn_scaled, verbose=0)
                    y_pred_test1 = model.predict(X_test1_nn_scaled, verbose=0)
                    y_pred_test2 = model.predict(X_test2_nn_scaled, verbose=0)

                    # AUC scores
                    auc_train = roc_auc_score(y_train, y_pred_train)
                    auc_test1 = roc_auc_score(y_test1, y_pred_test1)
                    auc_test2 = roc_auc_score(y_test2, y_pred_test2)

                    # Save result
                    result = {
                        '# HL': hl,
                        '# Node': n,
                        'Activation Function': act,
                        'Dropout': f"{int(dr * 100)}%",
                        'Batch Size': bs,
                        'AUC Train': round(auc_train, 4),
                        'AUC Test 1': round(auc_test1, 4),
                        'AUC Test 2': round(auc_test2, 4)
                    }

                    results.append(result)

                    # Save after each iteration
                    pd.DataFrame(results).to_csv("nn_grid_search_results.csv", index=False)

print("✅ Neural Network Grid Search Completed.")
````

---

### 🎯 Why Test 2 Matters More Than Test 1

In your project, you split the data like this:
- **Train (70%)**: Used to **fit** the model  
- **Test 1 (15%)**: Used to **tune** model parameters (like grid search)  
- **Test 2 (15%)**: Held out until the **very end** to measure **true performance**

---

### 🔍 The Roles of Each Set

| Set       | Purpose                                     | When It's Used        |
|-----------|---------------------------------------------|------------------------|
| **Train** | To train the model (learn patterns)         | Always                 |
| **Test 1**| To tune hyperparameters (e.g., grid search) | During optimization    |
| **Test 2**| Final evaluation of generalization ability  | Only after tuning done ✅ |

---

### 🧠 Why We Trust Test 2 the Most

- Test 1 **has been "seen" indirectly** — you've run multiple models and picked the best based on Test 1.
- Test 2 is **untouched and unseen** — like new data you'd face in real-world deployment.
- So, **Test 2 simulates actual future performance**. If a model does well here, it's not just lucky or overfitted.

---

### 📌 TL;DR

> ✅ **Test 2 is your gold-standard checkpoint** for generalization.  
> If a model performs well on Test 2, it's likely to perform well on unseen, real-world data.

````python
def evaluate_strategy(data, target_col, pd_col, balance_col, spend_col, threshold):
    # Filter accepted customers (low PD)
    accepted = data[data[pd_col] < threshold].copy()

    # Drop rows with missing spend/balance
    accepted = accepted.dropna(subset=[balance_col, spend_col])

    # Calculate default rate
    default_rate = accepted[target_col].mean()

    # Monthly revenue: balance × 0.02 + spend × 0.001
    accepted['monthly_revenue'] = accepted[balance_col] * 0.02 + accepted[spend_col] * 0.001

    # 12-month expected revenue (zero if they defaulted)
    accepted['expected_revenue'] = accepted.apply(
        lambda row: row['monthly_revenue'] * 12 if row[target_col] == 0 else 0,
        axis=1
    )

    # Return default rate and total revenue
    return default_rate, accepted['expected_revenue'].sum()
````

````python
import numpy as np
import pandas as pd

# Step 1: Strategy evaluation function
def evaluate_strategy(data, target_col, pd_col, balance_col, spend_col, threshold):
    accepted = data[data[pd_col] < threshold].copy()
    accepted = accepted.dropna(subset=[balance_col, spend_col])
    default_rate = accepted[target_col].mean()
    accepted['monthly_revenue'] = accepted[balance_col] * 0.02 + accepted[spend_col] * 0.001
    accepted['expected_revenue'] = accepted.apply(
        lambda row: row['monthly_revenue'] * 12 if row[target_col] == 0 else 0,
        axis=1
    )
    return len(accepted), default_rate, accepted['expected_revenue'].sum()

# Step 2: Define thresholds
thresholds = np.arange(0.1, 1.1, 0.1)

# Step 3: Loop through thresholds for each dataset
results_table = []

for t in thresholds:
    train_vals = evaluate_strategy(df_train, 'target', 'pd_xgb', 'B_40_Ave_6', 'S_3_Ave_6', t)
    test1_vals = evaluate_strategy(df_test1, 'target', 'pd_xgb', 'B_40_Ave_6', 'S_3_Ave_6', t)
    test2_vals = evaluate_strategy(df_test2, 'target', 'pd_xgb', 'B_40_Ave_6', 'S_3_Ave_6', t)

    results_table.append({
        "Threshold": round(t, 2),
        "Train # Total": train_vals[0],
        "Train Default Rate": round(train_vals[1], 4),
        "Train Revenue": round(train_vals[2], 2),
        "Test1 # Total": test1_vals[0],
        "Test1 Default Rate": round(test1_vals[1], 4),
        "Test1 Revenue": round(test1_vals[2], 2),
        "Test2 # Total": test2_vals[0],
        "Test2 Default Rate": round(test2_vals[1], 4),
        "Test2 Revenue": round(test2_vals[2], 2),
    })

# Step 4: Create DataFrame and display it
strategy_df = pd.DataFrame(results_table)
display(strategy_df)

# Optional: Save to CSV
strategy_df.to_csv("strategy_summary_table.csv", index=False)
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Load the strategy summary table
strategy_df = pd.read_csv("strategy_summary_table.csv")

# Plotting function for a single dataset
def plot_strategy(data, dataset_name):
    fig, ax1 = plt.subplots()

    ax1.set_xlabel('Threshold')
    ax1.set_ylabel('Expected Revenue ($)', color='tab:blue')
    ax1.plot(data['Threshold'], data[f'{dataset_name} Revenue'], color='tab:blue', label='Expected Revenue')
    ax1.tick_params(axis='y', labelcolor='tab:blue')

    ax2 = ax1.twinx()
    ax2.set_ylabel('Default Rate (%)', color='tab:red')
    ax2.plot(data['Threshold'], data[f'{dataset_name} Default Rate'] * 100, color='tab:red', linestyle='--', label='Default Rate')
    ax2.tick_params(axis='y', labelcolor='tab:red')

    plt.title(f"{dataset_name} Strategy: Revenue vs Default Rate")
    fig.tight_layout()
    plt.grid(True)
    plt.show()

# Generate charts
plot_strategy(strategy_df, "Train")
plot_strategy(strategy_df, "Test1")
plot_strategy(strategy_df, "Test2")
````
